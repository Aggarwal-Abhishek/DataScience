### Unsupervised Clustering Algorithm
Steps
1. Randomly assign each observation to a cluster
2. Find the centroid of each cluster
3. Reassing data points to the cluster whose centroid is closest
4. Calculate new centroid of each cluster
5. and Repeat ...

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Data

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

### Building Model

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
model.fit(data[0])
```

```python
print(model.cluster_centers_)
print(model.labels_)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('Prediction')
ax1.scatter(data[0][:,0],data[0][:,1],c=model.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(model.labels_, data[1]))
```

```python
print(confusion_matrix(model.labels_, data[1]))
```
